## Distplots

- Distplots are used to do univariate analysis on a continuous/numerical data.
- Distplots combine three things in one graph
  - Histogram
  - Kernel Density Estimation(PDF)
  - Rugplot

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic/train.csv')
df.head()
```

```python
df.shape
```

```python
df['Age'].isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Age'].isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(df['Age'])
plt.show()
```

### Plotting only the Histogram

Bins in a histogram are like containers that group together similar data values. Imagine you have a bunch of numbers, and you want to see how many fall into different ranges, like counting how many people fall into different age groups. Bins help organize these numbers into manageable groups so you can easily see patterns and trends in the data. They're like sorting your socks into different piles based on their colors.

```python
sns.distplot(df['Age'],bins=10, kde=False)
plt.show()
```

Range of y-axis is changed. in previous(distplot) shows the probability=0.1 of person at age=30 but here show number

### Plotting only KDE

```python
sns.distplot(df['Age'], hist=False)
plt.show()

# kde represent the pdf
# in our case the prob of finding the guy in age group of 20 to 40 is more
```

### Plotting only the Rugplot

```python
sns.distplot(df['Age'], hist=False, kde=False, rug=True)
plt.show()
```

- in this case we plot verticle line for each observation 
- for higher density -> thick line
- it tells in the range 60 to 80 the points are more tilted towards 60

## HOW TO USE DISTPLOT

```python
sns.distplot(df[df['Survived']==0]['Age'], color='red')
sns.distplot(df[df['Survived']==1]['Age'], color='green')
```

- age 0-10(child) are survived more
- age 30-40(adult) are not survived more
- age 50-60(old) are survived more
