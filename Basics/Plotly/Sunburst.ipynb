- A `sunburst` plot is a type of circular, multilevel plot that is useful for visualizing hierarchical data. It is similar to a pie chart but with multiple layers, where each layer represents a level in the hierarchy. Each segment of the circle represents a category or subcategory, with its size proportional to the value it represents.

**Features**
- `Center`: Represents the top of the hierarchy or the root.
- `Layers`: Represent different levels of the hierarchy. Moving outward from the center, each level represents subcategories or further breakdowns.
- `Colors and Sizes`: Segments are often colored to represent different categories, and their size corresponds to the value or frequency of each category.

**Use Case:**
- A `sunburst` plot is often used when you want to show how different parts contribute to the whole across multiple hierarchical levels. For example, you might use it to show sales data broken down by region, country, and city.
